```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def last_appearance(arr, x):
    if x in arr:
        return arr[::-1].index(x) + 1
    
    return 0
```

```python
last_appearance([0,0],1)
```

```python
N = 100000
```

```python
num = 0
vals = []

for _ in range(N):
    vals.append(num)
    num = last_appearance(vals[:-1], num)
```

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.plot(range(N),vals)

ax.set_title('Van Eck Sequence')

plt.show()
```
