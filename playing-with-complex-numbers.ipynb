```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
delta = 0.03
theta = 0.1
max_exp = 400
```

```python
def get_reals_imags(complex_num_list):
    reals = [complex(num).real for num in complex_num_list]
    imags = [complex(num).imag for num in complex_num_list]
    return reals, imags
```

```python
def draw_coords(ax):
    ax.axhline(y=0, color='k', lw=0.5)
    ax.axvline(x=0, color='k', lw=0.5)
```

Complex numbers **inside**, **over** and **outside** the unit circle:

```python
inside_UC = complex(math.cos(theta), math.sin(theta) - delta)
over_UC = complex(math.cos(theta), math.sin(theta))
outside_UC = complex(math.cos(theta), math.sin(theta) + delta)
```

```python
insides = [inside_UC**n for n in range(1, max_exp+1)]
overs = [over_UC**n for n in range(1, max_exp+1)]
outsides = [outside_UC**n for n in range(1, max_exp+1)]
```

```python
fig, ax = plt.subplots(figsize=(9,9))

fig.suptitle('Exponentiation of Complex Numbers', fontsize=20)
ax.set_title(f'$z_n = (z_0)^n \ :\ n = 1, 2, ..., {max_exp}; \ \\theta = {theta}; \ \delta = {delta}$', fontsize=16)

draw_coords(ax)

ax.plot(*get_reals_imags(insides), color='indigo', lw=0.50,
        label=r'$z_0 = \cos\ \theta + i\ (\sin\ \theta - \delta) $')
ax.plot(*get_reals_imags(overs), color='mediumorchid', lw=1.00,
        label=r'$z_0 = \cos\ \theta + i\ \sin\ \theta $')
ax.plot(*get_reals_imags(outsides), color='violet', lw=0.50,
        label=r'$z_0 = \cos\ \theta + i\ (\sin\ \theta + \delta) $')

ax.set_xlabel('Real part', fontsize=14)
ax.set_ylabel('Imag. part', fontsize=14)

ax.legend(prop={'size': 12})
plt.show()
```

Now, consider $z = \bigg(1 + i \displaystyle\frac{\pi}{m} \bigg)^m$, where $i = \sqrt{-1}$.

Each series $Z(m)$ will be given by $Z(m) = \Bigg\{ \bigg(1 + i \displaystyle\frac{\pi}{m} \bigg)^j \Bigg\}, \quad j = 1,2,...,m$.

```python
base = lambda m: complex(1, math.pi/m)
```

```python
m_a, m_b, m_c = 1e1, 1e2, 1e3
```

```python
Z_a = [base(m_a)**j for j in range(1, int(m_a)+1)]
Z_b = [base(m_b)**j for j in range(1, int(m_b)+1)]
Z_c = [base(m_c)**j for j in range(1, int(m_c)+1)]
```

```python
fig, ax = plt.subplots(figsize=(9,9))

draw_coords(ax)

ax.plot(*get_reals_imags(Z_a), label=f'$Z({m_a:.0e})$', marker='.')
ax.plot(*get_reals_imags(Z_b), label=f'$Z({m_b:.0e})$')
ax.plot(*get_reals_imags(Z_c), label=f'$Z({m_c:.0e})$')

ax.axis('scaled')
ax.legend()
plt.show()
```
