```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
num_fixed_points = 3
num_random_points = 20000
x_lim, y_lim = 100, 100
```

```python
def get_midpoint(x1, y1, x2, y2):
    return (x1+x2)/2, (y1+y2)/2
```

#### Generating fixed points

```python
fixed_xs, fixed_ys = [], []

for _ in range(num_fixed_points):
    x = np.random.uniform(low=0.0, high=x_lim)
    y = np.random.uniform(low=0.0, high=y_lim)
    fixed_xs.append(x)
    fixed_ys.append(y)
```

#### Generating random points

```python
random_xs = [np.random.uniform(low=0.0, high=x_lim)]
random_ys = [np.random.uniform(low=0.0, high=y_lim)]

for _ in range(num_random_points):
    target = np.random.choice(a=num_fixed_points)
    x, y = get_midpoint(fixed_xs[target],fixed_ys[target],random_xs[-1],random_ys[-1])
    random_xs.append(x)
    random_ys.append(y)
```

#### Plotting all points

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(9,9))
ax.scatter(fixed_xs,fixed_ys,s=100,c='k')

ax.scatter(random_xs,random_ys,s=25,c='r',alpha=0.5)
```
