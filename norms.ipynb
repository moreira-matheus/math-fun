```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
num = 5
size = 100
vectors = [np.random.randint(low=1,high=1000,size=size) for _ in range(num)]
```

```python
ps = np.arange(1,5)
plt.figure(figsize=(12,8))
for vector in vectors:
    norms = [np.linalg.norm(vector,ord=p) for p in ps]
    plt.plot(norms)
```
