```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def is_prime(n):
    if n == 1 or n == 0:
        return False
    elif n == 2 or n == 3:
        return True

    half = n//2 + 1
    for i in range(2,half):
        if (n % i) == 0:
            return False

    return True
```

```python
def prime_generator(n):
    for i in range(n):
        if is_prime(i):
            yield i
```

```python
def func_of_prime(prime):
    bin_form = bin(prime)[2:]
    rev_form = bin_form[::-1]

    return prime - int(rev_form, base=2)
```

```python
N = 50000
```

```python
xs = [p for p in prime_generator(N)]
```

```python
ys = [func_of_prime(x) for x in xs]
```

```python
fig, ax = plt.subplots(figsize=(10,8))

ax.scatter(xs, ys)

plt.show()
```
