#### State space:
$$
\vec{x} =
\begin{bmatrix}
x_0 \\
x_1 \\
x_2
\end{bmatrix}
$$

where $x_1$ is a proxy for deformation $s$, and $x_2$, for velocity $v$.

#### ODE:
$$
\dot{\vec{x}} =
\begin{bmatrix}
0 \\
x_2 \\
g - \frac{\beta}{m} x_2 - \frac{k}{m} x_1
\end{bmatrix}
$$

#### Initial conditions:

$$
\vec{x}(t=0) =
\begin{bmatrix}
1 \\
-1 \\
-2
\end{bmatrix}
\quad\quad
\Delta t = 0.1
$$

#### Update rule (Forward Euler):

$$
\vec{x} (t + \Delta t) = \vec{x}(t) + \Delta t \cdot \dot{\vec{x}} (t)
$$

##### Rewriting...

$$
\vec{x} (t + \Delta t) = \vec{x}(t) + \Delta t \cdot \bigg(M \cdot \vec{x} (t)\bigg)
$$

where:

$$
M =
\begin{bmatrix}
0 & 0 & 0 \\
0 & 0 & 1 \\
g & -\frac{k}{m} & -\frac{\beta}{m}
\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
g = 0.0
k = 2.0
m = 0.5
beta = 0.0
```

```python
M = np.array([[0, 0, 0],
              [0, 0, 1],
              [g, -k/m, -beta/m]])
```

```python
X0 = np.array([1, -1, -2]).T
```

```python
delta_t = 0.1
N = 2000
xs, vs = [X0[1]], [X0[2]]
```

```python
def derivative(X):
    global M
    return M.dot(X)
```

```python
def forward_euler(X):
    global delta_t
    return X + delta_t * derivative(X)
```

```python
for _ in range(N):
    X0 = forward_euler(X0)
    xs.append(X0[1])
    vs.append(X0[2])
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10,8))

ax.set_title('Simple Harmonic Oscillator: Forward Euler', fontsize=20)
ax.set_xlabel('Deformation ($s$)', fontsize=15)
ax.set_ylabel('Velocity ($v$)', fontsize=15)

ax.plot(xs,vs)
plt.show()
```
